```python
import torch
import matplotlib.pyplot as plt

seed = 100
path = './seq2seq/record/' + str(seed) + '/info_dict'
info_dict = torch.load(path)
train_loss_list = info_dict['train_loss_list']
validate_loss_list = info_dict['validate_loss_list']
train_acc_list = info_dict['train_acc_list']
validate_acc_list = info_dict['validate_acc_list']
step_finished = info_dict['step']
epoch = len(train_loss_list)

epochs = []
for i in range(epoch):
    epochs.append(i+1)
```

```python
num_hiddens = 128
warmup_steps = 4000
factor = 1
step_list = []
lrate_list = []

for i in range(step_finished):
    step_list.append(i+1)
    lrate = factor * num_hiddens ** (-0.5) * \
        min((i+1) ** (-0.5), (i+1) * warmup_steps ** (-1.5))
    lrate_list.append(lrate)
    
plt.plot(step_list,lrate_list)
plt.grid()
plt.show()
```

```python
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(epochs, train_loss_list, label='Training Loss')
plt.plot(epochs, validate_loss_list, label='Validatation Loss')
plt.title('Training and Validation Loss')
plt.grid()
plt.legend()
```

```python
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.plot(epochs, train_acc_list, label='Training Acc')
plt.plot(epochs, validate_acc_list, label='Validation Acc')
plt.title('Training and Validation Accuracy')
plt.grid()
plt.legend()
```
